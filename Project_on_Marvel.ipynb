```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
marvel = pd.read_csv('super_hero.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment_x'].value_counts()
```

```python
marvel[marvel['Alignment_x']=='good']
```

```python
good = marvel[['Name','Alignment_x','Speed','Power','Total','Intelligence_x','Combat']]
```

```python
good[good['Alignment_x']=='good']
good.head()
```

```python
for col in marvel.columns:
  print(col)
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
max_power = good[good['Power']==100]
```

```python
max_power.shape
```

```python
max_power
```

```python
max_power.sort_values(by=['Total'],ascending=False).head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(max_power['Name'])[:5],list(max_power['Total'])[:5],color=['green','yellow','orange','red','pink'])
plt.show()
```

```python
bad = good[good['Alignment_x']=='bad']
bad.head()
```

```python
bad.sort_values(by=['Speed'],ascending=False).head()
```

```python
bad.sort_values(by=['Intelligence_x'],ascending=False).head()
```

```python
bad.sort_values(by=['Total'],ascending=False).head()
```

```python
bad_power = bad.sort_values(by=['Total'],ascending=False)
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(bad_power['Name'])[:5],list(bad_power['Total'])[:5],color=['pink','red','orange','yellow','green'])
plt.show()
```

# Making a histogram for speed for Good superheros:

```python
plt.figure(figsize=(10,5))
plt.hist(good['Speed'])
plt.title('Speed of Good Superheros')
plt.xlabel('Speed')
plt.ylabel('Frequency')
plt.show()
```

# Making a histogram for speed for Bad superheros

```python
plt.figure(figsize=(10,5))
plt.hist(bad['Speed'])
plt.title('Speed of Good Superheros')
plt.xlabel('Speed')
plt.ylabel('Frequency')
plt.show()
```

# Making a histogram for Combat for supervillains


```python
plt.figure(figsize=(10,5))
plt.hist(bad['Combat'])
plt.title('Distribution of Combat')
plt.xlabel('Combat')
plt.show()
```
